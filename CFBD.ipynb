# Model

```python
import tensorflow as tf
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import AdaBoostClassifier
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis

from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
```

```python
y_values = final_df['home_win']
final_df = final_df.drop(['home_win'],axis=1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(final_df, y_values, test_size=0.33, random_state=42)
```

```python
X_train.head()
```

# Random Forest Model

```python
rf_mod = RandomForestClassifier(n_estimators=500, max_depth=10, bootstrap=True, random_state=42)

rf_mod.fit(X_train,y_train)
```

```python
rf_cv = cross_val_score(rf_mod,X_test,y_test,cv=7)

print(np.mean(rf_cv))
```

# Logistic Regression Model

```python
log_mod = LogisticRegression(random_state=42,solver='liblinear',max_iter=10000)

log_mod.fit(X_train,y_train)
```

```python
log_cv = cross_val_score(log_mod,X_test,y_test,cv=10)

print(np.mean(log_cv))
```

# Naive Bayes Model

```python
gnb = GaussianNB()

gnb.fit(X_train,y_train)
```

```python
gnb_cv = cross_val_score(gnb,X_test,y_test,cv=10)

print(np.mean(gnb_cv))
```
